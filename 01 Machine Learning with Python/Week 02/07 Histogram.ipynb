```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array(
    [
        3, 3, 5, 6,
        7, 7, 8, 9,
        9, 10, 10, 10,
        11, 12, 12, 14,
        15, 16, 17, 18,
        19, 19, 20
    ]
)
```

```python
plt.hist(data, bins=4, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency');
```
