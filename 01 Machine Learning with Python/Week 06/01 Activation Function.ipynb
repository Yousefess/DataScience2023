```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def s(x):
    return 1 / (1 + np.exp(-x))
```

```python
X = np.linspace(-10, 10, 50)
X
```

```python
plt.plot(X, s(X),'b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid()
plt.title('Sigmoid Function')
plt.text(2, 0.7, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=20 , color='g');
```

### Relu

```python
def ReLU(x):
    return np.maximum(0.0, x)
```

```python
X = np.linspace(-10, 10, 50)
plt.plot(X, ReLU(X),'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('ReLU Function')
plt.grid()
plt.text(-9, 6, r'$ReLU(x)=max(0.0, x)$', fontsize=20);
```
